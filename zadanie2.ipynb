**2.1.** Wygnerować listę 100 losowych liczb a potem wypisać ją od końca.

```python
import random
lista1 = []
N = 100

print("Wygenerowano listę:")
for i in range(N):
    x = random.randrange(0,N)
    print(x)
    lista1.append(x)
    
print("Lista od tyłu:")
lista1[::-1]
```

**2.2** Napisać funkcję, która zwraca 4 różne poziomy w zależności od wysokości temperatury

```python
def opis_temperatury():
    while True:
        temp = int(input("Podaj temperature: "))
        if temp < 0:
            print("bardzo zimno")
        elif 0 <= temp < 20:
            print("zimno")
        elif 20 <= temp < 25:
            print("cieplo")
        elif 25 <= temp:
            print("bardzo cieplo")
            
opis_temperatury()
```

**2.3** Zsumować z listy tylko te elementy, które są liczbami całkowitymi.

```python
x = 4
y = '2'
lista = ['a', 4, 2, [1, 2], -3, ['3', 4], -2, x, y]

def suma_calkowitych(lista):
    suma = 0
    for el in lista:
        if type(el) == int:
            suma += el
    return suma

suma_calkowitych(lista)
```
