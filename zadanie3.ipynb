Zadanie: Śledzenie Spędzonego Czasu

Stwórz program, który pozwala śledzić czas spędzony na różnych aktywnościach. Użyj słownika do przechowywania danych o aktywnościach oraz listy do śledzenia czasu spędzonego na każdej z nich. Program powinien mieć trzy główne funkcje:
* Dodaj Aktywność: Użytkownik może dodać nową aktywność do śledzenia. Jeśli aktywność już istnieje, program powinien tylko zaktualizować ją o nowy czas. Struktura danych może wyglądać tak:
    {
    'Aktywność 1': [czas1, czas2, czas3, ...],
    'Aktywność 2': [czas1, czas2, czas3, ...],
    ...
}
* Pokaż Czas: Użytkownik może wybrać aktywność i zobaczyć, ile łącznie czasu spędził na niej.
* Pokaż Top Aktywności: Program powinien pokazać top 3 aktywności, na których użytkownik spędził najwięcej czasu.

```python
dic = {} #przechowuje aktywności i ich czasy
dic_czasy = {} #przechowuje zsumowane czasy danych aktywności

print("Dodaj Aktywność - A")
print("Pokaż Czas - C")
print("Pokaż Top Aktywności - T\n")

while True:
    operacja = input("Wybierz rodzaj (literę) operacji: ")
    
    if operacja == 'A':
        a = input("Nazwa aktywności: ")
        c = int(input("Czas: "))
        if a in dic.keys():
            dic[a].append(c)
            act_suma = dic_czasy[a] + c
            dic_czasy.update({a : act_suma})
        else:
            dic[a] = [c]
            dic_czasy[a] = c
        print("--------dodano--------\n\n")
        
    elif operacja == 'C':
        nazwa = input("Podaj nazwę aktywności: ")
        if nazwa in dic.keys():
            print("Łączny czas czynności ", nazwa, ": ", dic_czasy[nazwa])
        else:
            print("Nie ma takiej aktywności")
        print("----------------------\n\n")
        
    elif operacja == 'T':
        #print(dic)
        #print(dic_czasy)
        lista = sorted(dic_czasy, key=dic_czasy.get) #sortujemy klucze słownika względem wartości
        print("Top 3 aktywności z najdłuższymi czasami:")
        n = len(lista)
        if n == 0:
            print("Nie dodano żadnych aktywności")
        elif n == 1:
            print("Mniej niż 3 aktywności")
            print("1. ", lista[n - 1], ": ", dic_czasy[lista[n - 1]])
        elif n == 2:
            print("Mniej niż 3 aktywności")
            for i in range(2):
                print(i + 1, ". ", lista[n - i - 1], ": ", dic_czasy[lista[n - i - 1]])
        else:
            for i in range(3):
                print(i + 1, ". ", lista[n - i - 1], ": ", dic_czasy[lista[n - i - 1]])
        print("----------------------\n\n")
        
    else:
        print("Błąd")
```
